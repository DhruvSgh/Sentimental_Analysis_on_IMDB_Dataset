```python
import pandas as pd
import numpy as np
import csv
from sklearn.feature_extraction.text import CountVectorizer
import nltk
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df= pd.read_csv('IMDB Dataset.csv')
#df= pd.read_csv('IMDB Dataset.csv',quoting=csv.QUOTE_NONE,on_bad_lines='skip')
#df = pd.read_csv(IMDB Dataset.csv, header = None, delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
```

```python
df.head(20)
```

```python
#vectorizer

tfidf= TfidfVectorizer(
    use_idf = True,
    norm = 'l2',
    smooth_idf=True
)
y=df.sentiment.values
X=tfidf.fit_transform(df['review'].values.astype('U'))
```

```python
#train and test

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.5,shuffle=False)
```

```python
df.dropna(inplace=True)
```

```python
import pickle
from sklearn.linear_model import LogisticRegressionCV
clf=LogisticRegressionCV(cv=5,
                         scoring='accuracy',
                         random_state = 0,
                         n_jobs=-1,
                         verbose=3,
                         max_iter=300).fit(X_train,y_train)
saved_model=open('saved_model.sav' ,'wb')
pickle.dump(clf, saved_model)
saved_model.close
```

```python
with open('saved_model.sav', 'rb') as f:
    saved_clf = pickle.load(f)

saved_clf.score(X_test,y_test)
```
